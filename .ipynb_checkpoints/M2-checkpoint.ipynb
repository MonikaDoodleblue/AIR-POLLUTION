# Data visualization:

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('city_day.csv')
```

```python
df = data.dropna()
```

```python
df.columns
```

```python
sns.countplot(x="AQI_Bucket", data=df, palette="husl")
```

```python
df['Xylene'].hist(figsize=(5,5), color='brown', alpha=0.7)
plt.xlabel('Xylene')
plt.ylabel('Count')
plt.title('Xylene and its Count')
```

```python
df['SO2'].hist(figsize=(5,5), color='yellow', alpha=1)
plt.xlabel('SO2')
plt.ylabel('Count')
plt.title('SO2 and its Count')
```

```python
def piechart(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(8,8), autopct='%1.2f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 15)

piechart(df, 'AQI_Bucket')
```

```python
g = sns.catplot("AQI_Bucket", data=df, kind='count', hue='City', margin_titles=False)
g.set_ylabels('Number of datas')
g.fig.set_figheight(5)
g.fig.set_figwidth(20)
```

```python
sns.scatterplot(df.Xylene, df.PM10)
```

```python
sns.pairplot(df)
```

```python
fig, ax = plt.subplots(figsize=(25,12))
sns.heatmap(df.corr(),annot = True, ax=ax)
plt.show()
```
